```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('odev_tenis.csv')
```

```python
data
```

```python
data2 = data.apply(preprocessing.LabelEncoder().fit_transform)
```

```python
c = pd.DataFrame(data2.iloc[:,:1])
ohe = preprocessing.OneHotEncoder()
c = ohe.fit_transform(c).toarray()
print(c)
```

```python
weather = pd.DataFrame(data= c , index = range(14), columns = ['o','r','s'])
```

```python
weather
```

```python
last_data = pd.concat([weather,data.iloc[:,1:3]],axis=1)
```

```python
last_data
```

```python
last_data = pd.concat([data2.iloc[:,-2:],last_data],axis=1)
```

```python
last_data
```

```python
x_train, x_test, y_train, y_test = train_test_split(last_data.iloc[:,:-1],last_data.iloc[:,-1:],test_size=0.33,random_state=0)
regressor = LinearRegression()
regressor.fit(x_train,y_train)
y_pred = pd.DataFrame(regressor.predict(x_test))
```
