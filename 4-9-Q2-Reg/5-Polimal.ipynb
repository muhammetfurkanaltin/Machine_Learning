# 5.1 Polimal Regresyon 
### Basit dogrusal regreskyon ->

    y = a + βx
    yᵢ = a + βxᵢ + εᵢ
    Satış = a + b(ay) + e

y: Bağımlı değişken (örneğin, satış) - 
x: Bağımsız değişken (örneğin, ay) - 
a: Sabit terim (y ekseni kesişimi) - 
β: Eğim (bağımsız değişkenin bağımlı değişken üzerindeki etkisi) - 
ε: Hata terimi


### Çoklu Dogrusal Regresyon ->

    y = β₀ + β₁x₁ + β₂x₂ + β₃x₃ + ε
    Boy = a + b(kilo) + c(yaş) + d(ayakkabı no) + e

y: Bağımlı değişken (örneğin, boy) - 
x₁, x₂, x₃: Bağımsız değişkenler (örneğin, kilo, yaş, ayakkabı numarası) - 
β₀: Sabit terim - 
β₁, β₂, β₃: Bağımsız değişkenlerin etkilerini gösteren katsayılar - 
ε: Hata terimi

### Polinomal Regresyon ->

    y = β₀ + β₁x + β₂x² + ... + βₕxʰ + ε
 
y: Bağımlı değişken - 
x: Bağımsız değişken - 
β₀, β₁, β₂, ..., βₕ: Polinomun katsayıları - 
h: Polinomun derecesi - 
ε: Hata terimi

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
data = pd.read_csv('maaslar.csv')
```

```python
data
```

```python
x = data.iloc[:,1:2]
y = data.iloc[:,2:]
X = x.values
Y = y.values
print(x,y)
```

#### Linear Regression

```python
lin_reg = LinearRegression()
lin_reg.fit(X,Y)
```

```python
plt.scatter(X,Y) # point
plt.plot(X,lin_reg.predict(X)) #line 
```

#### Polynomial Regression

from sklearn.preprocessing import PolynomialFeatures

```python
poly_reg = PolynomialFeatures(degree = 2)
x_poly = poly_reg.fit_transform(X)
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,Y)
```

```python
x_poly
```

```python
plt.scatter(X,Y)
plt.plot(X,lin_reg2.predict(poly_reg.fit_transform(X)))
```

```python
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(X)
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,Y)
```

```python
plt.scatter(X,Y)
plt.plot(X,lin_reg2.predict(poly_reg.fit_transform(X)))
```
